## Data Cleaning ##

```python
def clean_price(price):
    if isinstance(price,str):
        if '£' in price:
            price = price.replace('£','').replace('each est.','').replace('-','').strip()
        elif 'p' in price:
            price = '0.' + price.replace('p','').replace('each est.','').replace('-','').strip()

    return float(price)
```

```python
mid_df['price'] = mid_df['item-price'].apply(clean_price)
```

## specific Product ##

```python
Grapes_df = mid_df[
    mid_df['name'].str.contains('Grapes', case=False, na=False) & 
     (mid_df['category'].str.contains('everyday', case=False, na=False) | 
    ~mid_df['category'].str.contains('everyday', case=False, na=False))
]
```

```python
Grapes_remove = ['Waitrose Grapeseed Oil']

Grapes = Grapes_df[~Grapes_df['name'].str.contains('|'.join(Grapes_remove), case=False, na=False)]
```

```python
Grapes_plot = (
    Grapes.assign(new_price = lambda s: s['price'] / s['size'].str.replace('g','').astype('float') * 1000,
    price= lambda x: x['price'],
    size = lambda x: x['size'])
)



Grapes_plot[['name','price','size','new_price']]
```

```python
(
    ggplot(Grapes_plot, aes(x='name', y='new_price')) +
    geom_line(color='red', size=1.2) +
    labs(
        title='Price of Grapes ',
        x='Product Name',
        y='Price ($)'
    ) +  
    theme(
        axis_text_x=element_text(hjust=1),  
        plot_title=element_text(size=14, face='bold')  
    ) + ggsize(600, 800)
)
```

```python
(
    ggplot(Grapes_plot, aes(x='name', y='new_price')) +
    geom_point(color='blue', size=5) +
    labs(
        title='Price of Grapes by Name',
        x='Product Name',
        y='Price ($)'
    ) +  
    theme(
        axis_text_x=element_text(hjust=1),  
        plot_title=element_text(size=14, face='bold')  
    ) + ggsize(600, 800)
)
```
